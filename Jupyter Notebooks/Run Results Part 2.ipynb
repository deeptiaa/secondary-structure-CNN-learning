#### Skew of Dataset and Pearson's R 

```python
# Skew = 3 * (Mean – Median) / Standard Deviation

std = [0.943, 1.353, 2.205, 1.853, 1.981, 0.960, 0.568, 3.161, 0.399, 1.486]
mean = [3.069, -0.295, -2.055, -1.496, -1.162, -0.427, -0.377, -3.140, -0.173, -1.020]
median = [3.560, -0.251, -1.535, -0.611, -1.010, -0.448, -0.423, -2.614, -0.025, -0.412]

skew = []
for x in range(len(std)):
    skew_val = 3 * (mean[x] - median[x]) / std[x]
    skew.append(skew_val)

pr_no_noise = [0.237, 0.304, 0.191, 0.675, 0.485, 0.718, 0.450, 0.711, 0.530, 0.647]

print(stats.pearsonr(skew, pr_no_noise))
```
